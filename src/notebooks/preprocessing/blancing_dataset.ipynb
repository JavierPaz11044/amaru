```python
import pandas as pd
import amaru.utils as utils
import amaru.data_processing.clean_dataset as clean_dataset
import os
import dotenv
dotenv.load_dotenv()
path_config = os.getenv("PATH_CONFIG")
```

```python

cleaned_df = clean_dataset.CleanDataset(utils.get_random_path(), path_config).run()
# print(cleaned_df['dataset'].columns, len(cleaned_df['dataset'].columns))
df = cleaned_df['dataset']
print(df['timestamp'].dtype)
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp")
df =df.set_index("timestamp")
df.sort_index(inplace=True)
df_time_diffs = df.index.to_series().diff().dt.total_seconds()
print(df_time_diffs.value_counts())
print("=========")
df.drop_duplicates(keep="first", inplace=True, ignore_index=True)
#df_time_diffs = df.index.to_series().diff().dt.total_seconds()
print(df_time_diffs.value_counts())
```

```python
#TODO: Adjust the sequences to have a fixed time difference between them, for example, 1 second. This can be done by resampling the data and filling in missing values as needed.
def adjust_secuencences(df, column_timestamp, time_diff_threshold=1 ):
    df[column_timestamp] = pd.to_datetime(df[column_timestamp])
    df = df.sort_values(column_timestamp).set_index(column_timestamp)
    df[f"${column_timestamp}_window"] = df.index.floor(f"{time_diff_threshold}s")
    print(df)
    df_grouped = df.groupby(f"${column_timestamp}_window").mean()

    df2 = df_grouped.asfreq(freq=f"{time_diff_threshold}s", method='bfill')
    return df2

cleaned_df = clean_dataset.CleanDataset(utils.get_random_path(), path_config).run("label")
df = cleaned_df['dataset']
df_adjusted = adjust_secuencences(df, "timestamp", time_diff_threshold=1)
print(df_adjusted.index.to_series().diff().dt.total_seconds().value_counts())
```
