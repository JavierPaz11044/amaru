```python
# Clean data of nullable values, duplicates, and outliers. Also  Counter the number of rows and columns of the dataset. Finally, save the cleaned dataset in a new file.
import pandas as pd
import numpy as np

path = get_random_path()
def clean_dropna_dataset(path):
    df = pd.read_csv(path, index_col=False)
    original_shape = df.shape
    df_cleaned = df.dropna()
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
clean_dropna_dataset(path)
```

```python
import pandas as pd

def clean_drop_duplicates_dataset(path):
    df = pd.read_csv(path, index_col=False)
    original_shape = df.shape
    df_cleaned = df.drop_duplicates()
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
clean_drop_duplicates_dataset(path)
```

```python
import pandas as pd

path = get_random_path()

df =  pd.read_csv(path, index_col=False)
print(df.columns)

def format_columns_dataset(path, selected_columns=None):
    df = pd.read_csv(path, index_col=False)
    original_shape = df.shape
    df_cleaned = df.copy()
    df_cleaned.columns = [col.strip() for col in df_cleaned.columns]
    if selected_columns:
        df_cleaned = df_cleaned[selected_columns]
    df_cleaned.columns = [col.lower().replace(" ", "_").replace("/", "-") for col in df_cleaned.columns]
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
len(format_columns_dataset(path)['dataset'].columns)
```
