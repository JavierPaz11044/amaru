```python
from dotenv import load_dotenv
import os
import pathlib
load_dotenv()
os.getenv("BASE_PATH_DATA_RAW")
```

```python
import random
def get_random_path():
    base_path = os.getenv("BASE_PATH_DATA_RAW")
    raw_data_ = pathlib.Path(base_path)
    list_csv_files = list(raw_data_.rglob("*.csv"))
    position_random = random.randrange(0, len(list_csv_files)) 
    random_csv_file = list_csv_files[position_random]
    return random_csv_file
```

```python
# Clean data of nullable values, duplicates, and outliers. Also  Counter the number of rows and columns of the dataset. Finally, save the cleaned dataset in a new file.
import pandas as pd
import numpy as np

path = get_random_path()
def clean_dropna_dataset(path):
    df = pd.read_csv(path)
    original_shape = df.shape
    df_cleaned = df.dropna()
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
```

```python
import pandas as pd

def clean_drop_duplicates_dataset(path):
    df = pd.read_csv(path)
    original_shape = df.shape
    df_cleaned = df.drop_duplicates()
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
clean_drop_duplicates_dataset(path)
```

```python
import pandas as pd

path = get_random_path()

df =  pd.read_csv(path)
print(df.columns)

def format_columnas_dataset(path):
    df = pd.read_csv(path)
    original_shape = df.shape
    df_cleaned = df.copy()
    df_cleaned.columns = [col.strip().lower().replace(" ", "_").replace("/", "-") for col in df_cleaned.columns]
    return { "dataset": df_cleaned, 
             "category": "", 
             "metadata": {
                "num_rows": df_cleaned.shape[0], 
                "num_columns": df_cleaned.shape[1], 
                "original_num_rows": original_shape[0], 
                "original_num_columns": original_shape[1]
                }
        }
format_columnas_dataset(path)
```
